### 1) Tokenize the sequence and encode the label class

```python
# Custom tokenizing function for 'ATCGN'
def tokenize_sequence_numeric(sequence):
    nucleotide_mapping = {'A': 1, 'T': 2, 'C': 3, 'G': 4, 'N': 5}
    return [nucleotide_mapping[char] for char in sequence if char in 'ATCGN']
    
# Tokenize the sequences in columns starting with 'OG' and containing '_motif_'
for column in df.columns:
    if column.startswith('OG') and '_motif_' in column:
        df[column] = df[column].apply(tokenize_sequence_numeric)
```

```python
# Encode categorical string values in 'Subspecies'
le = LabelEncoder()
df['Subspecies'] = le.fit_transform(df['Subspecies'])
```

### 2) Extract the single nucleotide polymorphism (SNP) information for a single feature column

```python
# Find the column that starts with 'OG' and contains '_motif_'
feature_column = [column for column in df.columns if column.startswith('OG') and '_motif_' in column][0]
```

```python
# Find the index of the row with 'Row Header' column value of 'Typhimurium_1'
typhimurium_row_index = df.index[df['Row Header'] == 'Typhimurium_1'][0]

# Extract SNP positions for the specified row
sequence = df.loc[typhimurium_row_index, feature_column]
snp_positions = set()
for i in range(len(sequence)):
    if len(set([df.loc[index, feature_column][i] for index, _ in df.iterrows()])) > 1:
        snp_positions.add(i)

sorted_snp_positions = sorted(list(snp_positions))
```

```python
# New DataFrame to store SNP information
snp_df = pd.DataFrame(columns=['Representative_SNP'])

# Iterate through each row in the original dataframe to extract and store SNP information
for index, row in df.iterrows():
    sequence = row[feature_column]  # Get the sequence from the current row

    # Extract SNP information at specified positions in a sequential order
    SNP_info = ''.join([str(sequence[pos]) for pos in sorted_snp_positions])

    # Append the SNP information to the new DataFrame
    snp_df.loc[len(snp_df)] = [SNP_info]
```

### 3) Extract the single nucleotide polymorphism (SNP) information for all feature columns

```python
# Find the index of the row with 'Row Header' column value of 'Typhimurium_1'
typhimurium_row_index = df.index[df['Row Header'] == 'Typhimurium_1'][0]

# Iterate through columns starting with 'OG' and containing '_motif_'
for column in tqdm(df.columns, desc="Processing SNP Information"):
    if column.startswith('OG') and '_motif_' in column:
        feature_column = column
        representative_seq = df.loc[typhimurium_row_index, feature_column]  # Get the representative sequence from the specified row

        # Get the SNP positions for the current column
        snp_positions = set()
        for i in range(len(representative_seq)):
            if len(set([row[feature_column][i] for _, row in df.iterrows()])) > 1:  # Check for SNP positions
                snp_positions.add(i)

        sorted_snp_positions = sorted(list(snp_positions))

        # Extract and store SNP information for each row in the DataFrame
        snp_data = {'Representative_SNP': []}
        for index, row in df.iterrows():
            sequence = row[feature_column]  # Get the sequence from the current row

            # Extract SNP information at specified positions in a sequential order
            SNP_info = ''.join([str(sequence[pos]) for pos in sorted_snp_positions])

            snp_data['Representative_SNP'].append(SNP_info)

        # Store SNP information in a new column
        new_column_name = 'SNP_' + column
        df[new_column_name] = snp_data['Representative_SNP']

# Display the DataFrame with the added SNP information columns
print(df)
```
